```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("income.csv")
df
```

```python
df2 = pd.DataFrame({"Name":["Alex","Sam","Chips","Rachel","Ross"],"Age":[40,23,70,60,25],"Income($)":[120000,35000,200000,450000,100000]})
df2
```

```python
df.append(df2, ignore_index = True)
#append data twice
```

```python
df.drop_duplicates(subset ="Name",
                     keep = False, inplace = True)
#handle duplicate data
```

```python
df
```

```python
plt.scatter(df.Age, df["Income($)"])
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_predicted = km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['cluster'] = y_predicted
df
```

```python
df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
df3 = df[df.cluster == 2]
plt.scatter(df1.Age,df1['Income($)'], color = "green")
plt.scatter(df2.Age,df2['Income($)'], color = "blue")
plt.scatter(df3.Age,df3['Income($)'], color = "red")
plt.xlabel("Age")
plt.ylabel("Income($)")
#scaling is needed to be done as the two clusters are not correctly distributed
```

```python
scaler = MinMaxScaler()
df['Age'] = scaler.fit_transform(df['Age'].values.reshape(-1,1)) #typing mistake
df['Income($)'] = scaler.fit_transform(df['Income($)'].values.reshape(-1,1))
df
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Age','Income($)']])
df['cluster'] = y_predicted
```

```python
df
```

```python
df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
df3 = df[df.cluster == 2]
plt.scatter(df1.Age,df1['Income($)'], color = "green")
plt.scatter(df2.Age,df2['Income($)'], color = "blue")
plt.scatter(df3.Age,df3['Income($)'], color = "red")
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='yellow',marker='*',label='centroid')
plt.xlabel("Age")
plt.ylabel("Income($)")
```

```python
km.cluster_centers_
```

```python
#Utilizing Elbow method
sse = []
for k in range(1,10):
    km = KMeans(n_clusters=k)
    km.fit_predict(df[['Age','Income($)']])
    sse.append(km.inertia_) #intertia has SSE value for model
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of Squared Error')
plt.plot(range(1,10),sse)
```
